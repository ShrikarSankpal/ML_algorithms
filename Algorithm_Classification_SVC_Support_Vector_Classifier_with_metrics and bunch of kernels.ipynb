# Importing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# get data

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication',header=None)
data.columns = ['Variance', 'Skewness', 'Curtosis','Entropy','Class']
data.head()
```

# create target and variables

```python
x = data.drop('Class', axis = 1)
y = data['Class']
```

# Split data

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear',C=1.0)
model.fit(xtr,ytr)
```

# predictions

```python
y_pred = model.predict(xte)
```

# Result and metrics

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print(classification_report(yte,y_pred))
print(accuracy_score(yte,y_pred))
print(confusion_matrix(yte,y_pred))
```

```python
import numpy as np
randList=[]
for i in [10,20,30,40,50]:
    randList.append(np.random.randint(i))
randList
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report
import numpy as np

#dictionary to keep track of score and to calculate average
scoreDict={'linear':[],'poly':{1:[],2:[],3:[],4:[]},'rbf':[],'sigmoid':[]}


for randNum in randList:
    print("***************Random number: ",randNum,"\n")
    
    #splitting
    X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=randNum)
    
    #populate required values for model
    kernelLst=['linear', 'poly', 'rbf', 'sigmoid']
    polyDegree=[1,2,3,4]
    
    for kernel_ in kernelLst:
        if kernel_=='poly':
            for degree_ in polyDegree:
                model=SVC(kernel=kernel_,degree=degree_)
                model.fit(X_train, y_train)
                score=model.score(X_test,y_test)
                print(kernel_,"degree ",degree_," score: ",score)
                print("Confusion Matrix: \n",confusion_matrix(y_test,model.predict(X_test)))
                print("Classification Report: \n",classification_report(y_test,model.predict(X_test)))
                #send the score value to the dictionary
                scoreDict[kernel_][degree_].append(score)
        else:
            model=SVC(kernel=kernel_)
            model.fit(X_train, y_train)
            score=model.score(X_test,y_test)
            print(kernel_," score:  ",score)
            print("Confusion Matrix: \n",confusion_matrix(y_test,model.predict(X_test)))
            print("Classification Report: \n",classification_report(y_test,model.predict(X_test)))
            #send the score value to the dictionary
            scoreDict[kernel_].append(score)
        print("\n\n")
#print(scoreDict)
```
