# importing

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth
```

# get data

```python
dataset = pd.read_csv(r'datasets/groceriessmall.csv',header=None,sep = ',', names=range(20))
print((dataset.head()))
```

### convert data into list of lists after removing NA values

```python
dataset = dataset.apply(lambda x: list(x.dropna().values), axis=1).tolist()
print(dataset)
```

### Create a column for each category and make a crosstab

```python
tr = TransactionEncoder()
trenc = tr.fit(dataset)
trenc = trenc.transform(dataset)
print("true false list category wise: \n",trenc)
dataset1 = pd.DataFrame(trenc,columns=tr.columns_)
dataset1
```

# model build

* `min_support =transactions_where_item(s)_occur / total_transactions`.

```python
apriori(dataset1,min_support=0.3,use_colnames=True)
```

```python
fpgrowth(dataset1,min_support=0.4, use_colnames=True)
```
